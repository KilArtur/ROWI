# Посчтет GINI по всей выборке данных
Для расчета коэффициента Джини на основе предоставленных данных можно использовать AUC ROC-кривой. Коэффициент Джини в этом контексте связан с AUC следующим образом:

Gini=2×AUC−1

```python
df['Статус'].unique()
```

```python
# расчет Gini
df['Статус'] = df['Статус'].apply(lambda x: 0 if x == 'Хороший' else 1)

fpr, tpr, thresholds = roc_curve(df['Статус'], df['PD для 2-х'])
roc_auc = auc(fpr, tpr)

print(f'Коэффициент Джини: {2 * roc_auc - 1:.2f}')
```

**Вывод:**  Значение 0.57 можно считать удовлетворительным, что означает, что модель лучше случайной и обладает значительной предсказательной мощностью, хотя и не идеальной.

# Построение графика Roc-auc

```python
plt.figure(figsize=(12, 7))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC-кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.legend(loc="lower right")
plt.show()
```

**Вывод** Модель с AUC равным 0.78 может считаться довольно надежной во многих практических ситуациях, таких как кредитный скоринг, медицинская диагностика и т.д.

# Расчет коэффициента KS (Колмогорова-Смирнова)
Для расчета коэффициента KS нужно:  
- Отсортировать вероятности дефолта в порядке убывания
- Рассчитать кумулятивные доли положительных  и отрицательных классов
- Найти максимальное абсолютное отклонение между кумулятивными долями

```python
# расчет KS
probs = np.array(df['PD для 2-х'])
statuses = np.array(df['Статус'])

probs_sorted = np.sort(probs)[::-1]
statuses_sorted = statuses[np.argsort(probs)][::-1]

cumulative_positive_rates = np.cumsum(statuses_sorted) / np.sum(statuses)
cumulative_negative_rates = np.cumsum(1 - statuses_sorted) / np.sum(1 - statuses)

ks = np.max(np.abs(cumulative_positive_rates - cumulative_negative_rates))
print(f'Коэффициент KS: {ks:.2f}')
```

**Вывод::** Это высокое значение KS указывает на мощную способность модели классифицировать положительные и отрицательные случаи

# Построение графика функции распределения для двух классов

```python
sorted_probs = np.sort(probs)
sorted_statuses = statuses[np.argsort(probs)]

cumulative_default = np.cumsum(sorted_statuses)
cumulative_non_default = np.arange(1, len(probs) + 1) - cumulative_default

cumulative_default_ratio = cumulative_default / max(cumulative_default)
cumulative_non_default_ratio = cumulative_non_default / max(cumulative_non_default)

plt.figure(figsize=(10, 6))
plt.step(sorted_probs, cumulative_default_ratio, label='Кумулятивная доля отрицательных клиентов', where='post')
plt.step(sorted_probs, cumulative_non_default_ratio, label='Кумулятивная доля положительных клиентов', where='post')
plt.xlabel('Вероятность дефолта')
plt.ylabel('Кумулятивная доля')
plt.title('График функции распределения для двух классов')
plt.legend()
plt.grid(True)
plt.show()
```
